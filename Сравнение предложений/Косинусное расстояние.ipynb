```python
import numpy
import re
from scipy.spatial.distance import cosine
numpy.set_printoptions(threshold = numpy.nan)
L = []
D = {}
index = 0
with open("sentences.txt") as f:
    i = 0
    for line in f:
        print("{}. {}".format(i, line))
        print()
        L.append([x for x in re.split('[^a-z]', line.lower()) if x])
        i += 1
total_word = 0
for row in L:
    for el in row:
        total_word += 1
        if not el in D:
            D[el] = index
            index += 1
Matrix = numpy.zeros((len(L), len(D)), dtype = numpy.int8)
n_row = 0
max_sum = 0
for row in L:
    for el in row:
        Matrix[n_row, D[el]] += 1
        if Matrix[n_row, D[el]] > max_sum:
            max_sum = Matrix[n_row, D[el]]
    n_row += 1
n_row = 0
result_dict = {}
for vector in Matrix:
    result_dict[cosine(Matrix[0], vector)] = n_row
    n_row += 1
result_list = list(result_dict)
result_list.sort()
print("ТОП2 наиболее близких к нулевому предложению: {}, {}".format(result_dict[result_list[1]], result_dict[result_list[2]]))
```
